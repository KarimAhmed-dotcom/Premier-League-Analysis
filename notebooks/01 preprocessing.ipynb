### Data Preprocessing

```python
import pandas as pd
```

```python
df=pd.read_csv('../data//raw/premier_league.csv')
```

```python
df.head()
```

**transform date and time columns into datetime**

```python
df['Date']=pd.to_datetime(df['Date'],format='%Y-%m-%d',errors='coerce')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')
```

**delete Match [Report , Notes , xG_x , Attendance] columns**

```python
df.drop(columns=['Match Report','Notes','xG_x','Attendance','Comp'],axis=1,inplace=True)
```

**Rename xG_y column into XG**

```python
df=df.rename(columns={'xG_y':'XG'})
```

**fill 69 values of [G/SoT , npxG/Sh , G/Sh , SoT% columns]**

```python
mask=df['G/SoT'].isna()
```

```python
mask_df=df.loc[mask]
```

```python
df.loc[mask,'G/SoT']=mask_df['GF']/mask_df['SoT']
```

```python
#expalin why nan because we divide about zer0
df[['GF','SoT']][df['G/SoT'].isna()].iloc[:30]
```

```python
df.loc[df['G/SoT'].isna(),'G/SoT']=0
```

```python
df.loc[df['npxG/Sh'].isna(),'npxG/Sh']=0
```

```python
df.loc[df['G/Sh'].isna(),'G/Sh']=1
```

```python
df.loc[df['SoT%'].isna(),'SoT%']=0
```
